```python
# Importing Pandas Library
import pandas as pd

# reading file as a dataframe
df = pd.read_excel('Customer Call List.xlsx')

#Imported Data
df
```

```python
# Removing Duplicate Data
df = df.drop_duplicates()

# Removing useless columns
df = df.drop(columns = "Not_Useful_Column")

# Cleaning Last_Name Column
df["Last_Name"] = df["Last_Name"].str.strip("123/._")

# Cleaning Phone_Number Column
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','-')
df["Phone_Number"] = df["Phone_Number"].str.replace('N-a','')
df["Phone_Number"] = df["Phone_Number"].fillna('')

#spliting and creating new address columns
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',2,expand=True)

# Changing "Y" to "Yes" and "N" to "No" in Paying Customer & Do_Not_Contact Columns
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('Y','Yes')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Paying Customer'] = df['Paying Customer'].str.replace('N','No')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Y','Yes')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('N','No')

# Removing NaN Values in Paying Customer & Do_Not_Contact Columns
df['Paying Customer'] = df['Paying Customer'].str.replace('No/a','')
df['Do_Not_Contact'] = df['Do_Not_Contact'].fillna('')

# Removing null Values across the DataFrame
df = df.replace('N/a')
df = df.fillna('')

# Removing Customers who does not want to get called or doesn't have a Phone Number
for i in df.index:
    if df.loc[i, 'Phone_Number'] == '':
        df.drop(i, inplace=True)
    elif df.loc[i, 'Do_Not_Contact'] == 'Yes':
        df.drop(i, inplace=True)
        
# Removing Useless Columns
df = df.drop(columns="Zip_Code")
df = df.drop(columns="Address")

# Index Reseting
df = df.reset_index(drop=True)

# Arranging Columns
df.insert(4, 'Street_Address', df.pop('Street_Address'))
df.insert(5, 'State', df.pop('State'))

#Final Data
df
```
